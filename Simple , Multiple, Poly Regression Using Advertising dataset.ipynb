```
import numpy as np
import pandas as pd
```

```
a = np.array([10,25,30,45,38,56,45,43,200,250])
```

```
a.mean()
```

```
np.median(a)
```

```
q1= np.quantile(a,0.25)
q3= np.quantile(a,0.75)
print(q1,q3)
```

```
iqr = q3-q1
print(iqr)
```

```
lw= q1-(1.5*iqr)
uw= q3+(1.5*iqr)
print(lw)
print(uw)
```

```
# skewness
from scipy.stats import skew
```
