```
import pandas as pd
import numpy as np
```

```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
iris
```

```
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Species']=iris.target
```

```
df.head()
```

```
df.shape
```

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```
x_train,x_test , y_train,y_test = train_test_split(df.drop('Species',axis=1), df['Species'],test_size=0.2,random_state =1)
rfc = RandomForestClassifier(n_estimators=10,n_jobs=2,verbose=2)
rfc.fit(x_train, y_train)
```

```
rfc.score(x_test, y_test)
```

```
rfc.feature_importances_
```

```
feature_imp = pd.Series(rfc.feature_importances_, index=iris.feature_names).sort_values(ascending=False)
print(feature_imp)
```

```
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Species','sepal width (cm)'] , axis=1), df['Species'], test_size=0.2, random_state=1)
rfc = RandomForestClassifier(n_estimators=10, n_jobs=2, verbose=2)
rfc.fit(x_train, y_train)
rfc.score(x_test, y_test)
```
