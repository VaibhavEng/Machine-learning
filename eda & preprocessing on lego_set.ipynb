# PRE- PROCESSING

```
df['num_reviews'].fillna(df['num_reviews'].mean(),inplace=True)
```

```
df['play_star_rating'].fillna(df['play_star_rating'].mean(),inplace=True)
```

```
df['star_rating'].fillna(df['star_rating'].mean(),inplace=True)
```

```
df['review_difficulty'].fillna(df['review_difficulty'].mode()[0], inplace=True)
df['review_difficulty'].unique()
```

```
df['theme_name'].fillna(df['theme_name'].mode()[0], inplace=True)
df['theme_name'].unique()
```

```
df['val_star_rating'].fillna(df['val_star_rating'].mean(),inplace=True)
```

```
plt.figure(figsize=(7,8))
sns.boxplot(data=df,x="list_price",y="theme_name")
plt.show()
```

```
df_cat = df.select_dtypes(object)
df_num = df.select_dtypes(["float64"])
```

```
df_cat.head()
```

```
df_num.head()
```

```
# for skew  import library
from scipy.stats import skew
```

```
for col in df_num:
  print(col,":",skew(df_num[col]))
  sns.distplot(df_num[col])
  plt.show()
  print("-------------------------------------")
```

```
sns.heatmap(df_num.corr(), annot=True)
plt.show()
```

```
df_num['num_reviews'] = np.sqrt(df_num['num_reviews'])
```

```
print(skew(df_num['num_reviews']))
```

```
df_num['play_star_rating'] = np.log(df_num['play_star_rating'])
```

```
print(skew(df_num['play_star_rating']))
```

```
df_num['star_rating'] = np.log(df_num['star_rating'])
```

```
print(skew(df_num['star_rating']))
```

```
from sklearn.preprocessing import LabelEncoder
```

```
for col in df_cat:
    le = LabelEncoder()
    df_cat[col] = le.fit_transform(df_cat[col])
df_cat.head()
```

```
df_new = pd.concat([df_num, df_cat], axis=1)
df_new.head()
```

```
df_mm = df_new.copy()
df_ss = df_new.copy()
```

```
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```
for col in df_mm:
    mm = MinMaxScaler()
    df_mm[col] = mm.fit_transform(df_mm[[col]])
df_mm.head()
```

```
for col in df_ss:
    ss = StandardScaler()
    df_ss[col] = ss.fit_transform(df_ss[[col]])
df_ss.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```
x = df_new.drop('list_price',axis=1)
y = df_new['list_price']
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
lr= LinearRegression()
lr.fit(x_train,y_train)
y_hat = lr.predict(x_test)
print("MSE = {}\nR2={}".format(mean_squared_error(y_test,y_hat),r2_score(y_test, y_hat)))
```

```
x = df_mm.drop('list_price',axis=1)
y = df_mm['list_price']
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
lr= LinearRegression()
lr.fit(x_train,y_train)
y_hat = lr.predict(x_test)
print("MSE = {}\nR2={}".format(mean_squared_error(y_test,y_hat),r2_score(y_test, y_hat)))
```

```
x = df_ss.drop('list_price',axis=1)
y = df_ss['list_price']
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
lr= LinearRegression()
lr.fit(x_train,y_train)
y_hat = lr.predict(x_test)
print("MSE = {}\nR2={}".format(mean_squared_error(y_test,y_hat),r2_score(y_test, y_hat)))
```

```
# num_reviews,piece_count,theme_name,play_star_rating,'star_rating'
x = df_new[['num_reviews','piece_count','theme_name','play_star_rating','star_rating']]
y = df_new['list_price']
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
lr= LinearRegression()
lr.fit(x_train,y_train)
print("intercept ={}\nslope={}".format(lr.coef_,lr.intercept_))
y_hat = lr.predict(x_test)
print("MSE = {}\nR2={}".format(mean_squared_error(y_test,y_hat),r2_score(y_test, y_hat)))
```

```
df_new['ratings']=df_new['play_star_rating']*df_new['star_rating']
```

```
x = df_new[['num_reviews','piece_count','theme_name','ratings']]
y = df_new['list_price']
x_train,x_test , y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
lr= LinearRegression()
lr.fit(x_train,y_train)
print("intercept ={}\nslope={}".format(lr.coef_,lr.intercept_))
y_hat = lr.predict(x_test)
print("MSE = {}\nR2={}".format(mean_squared_error(y_test,y_hat),r2_score(y_test, y_hat)))
residuals = y_test - y_hat
sns.distplot(residuals)
plt.show()
sns.scatterplot(y_hat,residuals)
plt.show()
```

```
# Plotting Actual vs Predicted Values
sns.scatterplot(x=y_test, y=y_hat)
sns.lineplot(y_test, y_test,  color='red') 

plt.xlabel('actual  Price', fontsize = 15)
plt.ylabel('Prdict list price', fontsize = 15)
```

```
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test,y_hat))
print('MSE:', metrics.mean_squared_error(y_test, y_hat))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_hat)))
```
